```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# load one file from one session
data = np.loadtxt("../_readings_right_hand/12345-1/2.txt", delimiter = ",")
data.shape, data.dtype
```

```python
def draw(xlim):
    plt.figure(figsize=(14, 8))
    # draw EMG
    for chan in range(8):
        plt.subplot(9, 1, chan + 1)
        plt.plot(data[:, chan])
        plt.ylim(-150, 150)
        plt.xlim(xlim)
        plt.xticks([])
    # draw class/gesture label
    plt.subplot(9, 1, 9)
    plt.plot(data[:, 8], color = "orange")
    plt.xlim(xlim)
    
# draw whole graphs
draw(None)
# draw first 10 seconds
draw((0, 2000))
```
